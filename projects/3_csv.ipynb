# 處理 CSV 檔案

處理csv檔案，有兩種常見的方式。先講述第一種我們之前學過的，使用檔案讀取，然後使用split。第二種使用pandas創造dataframe.

1. read file and split
2. pandas

```python
# 檔案來源
# https://data.gov.tw/dataset/161693

with open('resources/11201結婚.csv', 'rb') as file_in:
    lines = file_in.readlines()
    lines = [e.decode('utf-8') for e in lines]
```

```python
for each in lines:
    # 用逗號來分隔
    parts = each.split(',')
    
    # do something
    
    # print
    print(parts[0], ' | ',parts[1])
    
    
```
